```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
i=[]
g=[]
g1=[]
g2=[]
g3=[]
a='రైతు'
a1='నిరసన'
a2='చట్టాల'
a3='పంజాబ్'
a4='దిల్లీ'
for l in range(2,99):
    url="https://www.bbc.com/telugu/topics/c5qvp16w7dnt/page/"
    url+=str(l)
    b=requests.get(url)
    
    c=BeautifulSoup(b.text,'lxml')
    
    for j in c.find_all('div',{'class':'gel-5/8@l'}):
        d=j.findAll('p',{'class':'lx-stream-related-story--summary qa-story-summary'})
        for q in d:
            #print(q.text)
            i.append(q.text)
    for j in c.find_all('time',{'class':'lx-stream-post__meta-time gs-u-align-middle gs-u-display-inline-block gs-u-mr0@m gs-u-mr gel-long-primer'}):
        d=j.findAll('span',{'class':'gs-u-vh qa-visually-hidden-meta'})
        for q in d:
            #print(q.text)
            g.append(q.text)
    for j in c.find_all('a',{'class':'qa-heading-link lx-stream-post__header-link'}):
        d=j.findAll('span',{'class':'lx-stream-post__header-text gs-u-align-middle'})
        for q in d:
            #print(q.text)
            g1.append(q.text)
    for j in c.find_all('div',{'class':'lx-stream-post__contributor-body gs-o-media__body'}):
        d=j.findAll('p',{'class':'qa-contributor-name lx-stream-post__contributor-name gel-long-primer gs-u-m0'})
        for q in d:
            #print(q.text)
            g2.append(q.text)
    for j in c.find_all('div',{'class':'lx-stream-post__contributor-body gs-o-media__body'}):
        d=j.findAll('p',{'class':'qa-contributor-role lx-stream-post__contributor-description gel-brevier gs-u-m0'})
        for q in d:
            #print(q.text)
            g3.append(q.text)
s=len(i)
pp=0
for q in range(s):
    if(((i[q].find(a))!=-1)and(((i[q].find(a1))!=-1) or (i[q].find(a2))!=-1 or ((i[q].find(a3))!=-1) or ((i[q].find(a4))!=-1)) ):
        pp+=1
        print(pp)
        print("Date:"+g[q])
        print("Headlines:"+g1[q])
        print("Article Writer:"+g2[q])
        print("Writer's Role:"+g3[q])
        print("News:"+i[q])
        print("----------(***)--------")
    
```
