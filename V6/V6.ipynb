```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import requests
from bs4 import BeautifulSoup
date=[]
links=[]
name=[]
qq=0
for kk in range(1,10):
    a='https://www.v6velugu.com/tag/farmers-protest/page/'
    a2=str(kk)
    a+=a2
    b=requests.get(a)
    c=BeautifulSoup(b.content,'lxml')

    for j in c.find_all('div',{'class':'dt-news-post-content'}):
        qq+=1
        print(qq)
        for i in j.find_all('span',{'class':'dt-news-post-date'}):
            date.append(i.text)
            print("Date:"+i.text)
        for link in j.find_all('a',title=True):
            name.append(link['title'])
            print('News:'+link['title'])
        for link in j.find_all('a',href=True):
            links.append(link['href'])
            print(link['href'])
        print("-------(**)-------")
        
```
